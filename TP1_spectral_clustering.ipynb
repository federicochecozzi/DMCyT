```python
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import cv2 as cv
```

```python
img = cv.imread('/content/logo2.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.title(img)
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
ret,th1 = cv.threshold(gray,0,255,cv.THRESH_BINARY)
plt.title(th1)
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.imshow(th1,'gray')
```

```python
th1.shape
```

```python
th1[0,0]
```

```python
list_pixel = []
for i in reversed(range(0,128)):
  for j in range(0,128):
    if th1[i,j] == 255:
      list_pixel.append([j,127-i])
array_pixel = np.array(list_pixel)
```

```python
array_pixel.shape
```

```python
sc = SpectralClustering(n_clusters=3).fit(array_pixel)
```

```python
labels = sc.labels_

plt.scatter(array_pixel[:,0], array_pixel[:,1], c=labels)
plt.show()  
```
